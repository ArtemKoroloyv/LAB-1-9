```python
def vivod_file(x, file):
    with open(file, "r") as file:
        if x == "Целиком":
            content = file.read()
            print(content)
        elif x == "По строкам":
            for line in file:
                print(line)
        else:
            print("Введите корректный тип чтения!")

vivod_file("Целиком","example.txt")
vivod_file("По строкам", "example.txt")
```

```python
def vivod_file(x, file):
    try:
        if x == "Целиком":
            with open(file, "r") as file:
                content = file.read()
                print(content)
        elif x == "По строкам":
            with open(file, "r") as file:
                for line in file:
                    print(line)
        else:
            print("Введите корректный тип чтения!")
    except FileNotFoundError:
       print("Файл не существует!")

vivod_file("Целиком","example.txt")
vivod_file("По строкам", "example.txt")
vivod_file("По строкам", "Not_found.txt")
```
